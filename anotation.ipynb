### Convertendo Tipos
//Converte para string
int a = Convert.ToInt32("10");
int a = Parse("10");
int a = int.Parse("10c"); // Isso vai gerar um erro, pois "10c" não é um número válido
Console.WriteLine(a);

Diferença entre ToInt32 e Parse
ToInt32 -> Valor pode ser null (Mais comum de usar)
Parse -> Valor nunca pode ser null

// Cast Implicito
int a = 5;
double b = a; // Converte implicitamente de int para double
Console.WriteLine(b);

// Converte para string
string inteiro = Convert.ToString(10);
Console.WriteLine(inteiro);

// Convertendo de maneira segura
string a = "15-";
int b = 0;
int.TryParse(a, out b); // Tenta converter a string para int, out é o parâmetro de saída
